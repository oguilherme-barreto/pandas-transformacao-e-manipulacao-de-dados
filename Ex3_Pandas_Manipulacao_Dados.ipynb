```python
import pandas as pd
import requests
import numpy as np
```

```python
url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
```

```python
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    dados = pd.DataFrame(data)
else:
    print(f"Erro: {response.status_code}")
dados
```

# Etapa 1

```python
dados = pd.json_normalize(dados['dados_locacao'])
dados
```

# Etapa 2
- Remover os dados em listas dentro do DataFrame;
- Verificar os tipos de dados;
- Identificar colunas numéricas;
- Transformar a coluna numérica para o tipo numérico.

```python
colunas_listas = ['datas_combinadas_pagamento', 'datas_de_pagamento', 'valor_aluguel']
```

```python
dados = dados.explode(colunas_listas, ignore_index=True)
dados
```

```python
dados.info()
```

```python
colunas_numericas = ['valor_aluguel']
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','.').replace('reais',''))
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].astype(np.float64)
```

```python
dados.info()
```

# Etapa 3
Manipule os textos na coluna apartamento para remover o texto (blocoAP) do DataFrame.

```python
dados['apartamento'] = dados['apartamento'].str.replace('(blocoAP)', '')
```
